```python
import pandas as pd
import numpy as np

# 1. إنشاء داتا تحاكي سلوك العملاء في الشراء
customer_data = {
    'Customer_ID': ['C-101', 'C-102', 'C-103', 'C-104', 'C-105', 'C-106', 'C-107', 'C-108'],
    'Customer_Name': ['Shimaa', 'Ahmed', 'Sara', 'Ali', 'Mona', 'Omar', 'Hoda', 'Youssef'],
    'Recency_Days': [2, 45, 1, 120, 5, 90, 3, 60],        # اشترى من كام يوم؟
    'Frequency_Count': [15, 2, 22, 1, 12, 3, 18, 4],     # عدد مرات الشراء
    'Total_Spend': [8500, 450, 12000, 150, 6200, 800, 9500, 1100] # إجمالي اللي دفعه
}

df_customers = pd.DataFrame(customer_data)
print("--- داتا العملاء وسلوكهم الشرائي ---")
print(df_customers)
```

```python
# دالة برمجية ذكية لتصنيف العميل بناءً على سلوكه
def assign_segment(row):
    if row['Recency_Days'] <= 5 and row['Frequency_Count'] >= 12 and row['Total_Spend'] >= 6000:
        return 'VIP Customer'
    elif row['Recency_Days'] <= 30 and row['Frequency_Count'] >= 4:
        return 'Loyal Customer'
    elif row['Recency_Days'] > 60 or row['Total_Spend'] < 500:
        return 'At Risk (⚠️ معرض للمغادرة)'
    else:
        return 'Regular Customer'

# تطبيق الدالة على الجدول لإنشاء عمود التصنيف الجديد
df_customers['Segment'] = df_customers.apply(assign_segment, axis=1)

print("--- نتيجة تقسيم وتصنيف العملاء الاحترافية ---")
print(df_customers[['Customer_Name', 'Recency_Days', 'Frequency_Count', 'Total_Spend', 'Segment']])
```
