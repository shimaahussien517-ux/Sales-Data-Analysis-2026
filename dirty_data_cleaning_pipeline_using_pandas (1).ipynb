```python
import pandas as pd
import numpy as np

# 1. تجهيز داتا "ملوثة" ومبهدلة لمحاكة الواقع
dirty_data = {
    'Transaction_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Branch': ['Assiut', 'Cairo', 'Assuit', 'Alex', 'cairo', 'Assiut', 'Cairo', 'Sohag', 'Assiut', 'Alex'],
    'Customer_Email': ['shimaa@gmail.com', '  cairo.user@yahoo.com ', 'SHIMAA@gmail.com', 'alex_eng@outlook.com', 'cairo.user@yahoo.com', 'SHIMAA@GMAIL.COM', np.nan, 'sohag.eng@gmail.com', 'shimaa@gmail.com', 'alex_eng@outlook.com'],
    'Sales_Amount': [5000, -1500, 4800, np.nan, 3200, 5200, 4100, 2900, -500, 3800],
    'Date': ['2026-01-01', '02/01/2026', '2026-01-02', '2026-01-03', '2026/01/03', np.nan, '2026-01-04', '2026-01-05', '2026-01-05', '2026-01-06']
}

df_dirty = pd.DataFrame(dirty_data)
print("--- الداتا المبهدلة قبل التنظيف ---")
print(df_dirty)
```

```python
# 1. توحيد طريقة الكتابة (تخيل كل الحروف Capitalize)
df_dirty['Branch'] = df_dirty['Branch'].str.capitalize()

# 2. تصحيح الغلطة الإملائية لفرع أسيوط
df_dirty['Branch'] = df_dirty['Branch'].replace('Assuit', 'Assiut')

print("--- عمود الفروع بعد التنظيف والتوحيد ---")
print(df_dirty[['Branch']])
```

```python
# 1. تحويل أي قيم سالبة في المبيعات إلى موجبة
df_dirty['Sales_Amount'] = df_dirty['Sales_Amount'].abs()

# 2. حساب متوسط المبيعات (بدون القيم المفقودة)
sales_mean = df_dirty['Sales_Amount'].mean()

# 3. ملء القيم المفقودة (NaN) بالمتوسط اللي حسبناه
df_dirty['Sales_Amount'] = df_dirty['Sales_Amount'].fillna(sales_mean)

print("--- عمود المبيعات بعد تطيير السوالب وملء الـ NaN ---")
print(df_dirty[['Sales_Amount']])
```

```python
# 1. إزالة المسافات الزائدة وتحويل الحروف كلها لـ سمول
df_dirty['Customer_Email'] = df_dirty['Customer_Email'].str.strip().str.lower()

# 2. ملء الإيميلات المفقودة بكلمة Unknown
df_dirty['Customer_Email'] = df_dirty['Customer_Email'].fillna('unknown')

print("--- عمود الإيميلات بعد التنظيف التام ---")
print(df_dirty[['Customer_Email']])
```

```python
# 1. تحويل عمود التواريخ إلى صيغة DateTime موحدة ذكياً
df_dirty['Date'] = pd.to_datetime(df_dirty['Date'], format='mixed')

# 2. ملء التاريخ المفقود بأول تاريخ في الداتا (أو تقدري تسيبيه)
most_frequent_date = df_dirty['Date'].mode()[0]
df_dirty['Date'] = df_dirty['Date'].fillna(most_frequent_date)

print("--- الجدول بالكامل بعد ما بقى نضيف وفلة 100% ---")
print(df_dirty)
print("\n--- أنواع البيانات للتأكد من تحويل التاريخ ---")
print(df_dirty.dtypes)
```

```python
# حساب إجمالي المبيعات لكل فرع بعد التنظيف
branch_summary = df_dirty.groupby('Branch')['Sales_Amount'].sum().sort_values(ascending=False)
print("\n🏆 ترتيب الفروع حسب إجمالي المبيعات بعد تنظيف الداتا:")
print(branch_summary)
```
