```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. بناء داتا تاريخية لمبيعات فرع أسيوط (الشهور الـ 6 الأولى من سنة 2026)
# Month_Num: ترتيب الشهر (1 تعني يناير، 2 فبراير، وهكذا)
# Marketing_Spend: الميزانية اللي صرفناها على الإعلانات في الشهر ده
# Sales_Amount: المبيعات الحقيقية اللي فرع أسيوط حققها
data_ml = {
    'Month_Num': [[1], [2], [3], [4], [5], [6]],
    'Marketing_Spend': [5000, 7000, 8000, 10000, 12000, 15000],
    'Sales_Amount': [15000, 22000, 25000, 31000, 37000, 46000]
}

# 2. استدعاء موديل الذكاء الاصطناعي
model = LinearRegression()

# 3. تدريب الموديل (Fit) -> هنا الموديل بيذاكر العلاقة بين الشهور والمبيعات
model.fit(data_ml['Month_Num'], data_ml['Sales_Amount'])

print("🎯 تم تدريب موديل الذكاء الاصطناعي بنجاح يا هندسة!")
print("الموديل دلوقتي جاهز يتوقع مبيعات أي شهر مستقبلي!")
```

```python
# طلب التنبؤ للشهر السابع والثامن
future_months = [[7], [8]]
predictions = model.predict(future_months)

print(f"🔮 مبيعات فرع أسيوط المتوقعة لشهر يوليو (شهر 7) هي: {predictions[0]:.2f} جنيه")
print(f"🔮 مبيعات فرع أسيوط المتوقعة لشهر أغسطس (شهر 8) هي: {predictions[1]:.2f} جنيه")
```
